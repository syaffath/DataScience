```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heartfail.csv")
df.head()
```

```python
#==============================================DATA VISUALIZATION============================================#
```

```python
#Age Distribution

Hist_Age = df['age'].plot(kind='hist', alpha=0.6, title='Age Distribution', bins=50) #density=True
Hist_Age.set_xlabel('Age')
```

```python
sns.distplot(df['age'], bins = 50, rug=True, kde_kws={"lw": 2, "label": "Age Distribution"})
```

- Age wise 40 to 80 the spread is High
<br>
- less than 40 age and higher than 80 age people are very low

```python
# Age VS Sex

Box_AgeSex = sns.boxplot(x=df['sex'], y=df['age'], palette="Set2")
Box_AgeSex = sns.swarmplot(x=df['sex'], y=df['age'], color=".25")
```

- Male = 1
<br>
- Female = 0

```python
# Percentage Sex vs Death Event
male = df[df["sex"]==1]
female = df[df["sex"]==0]

male_survived = len(male[df["DEATH_EVENT"]==0])
male_not = len(male[df["DEATH_EVENT"]==1])
female_survived = len(female[df["DEATH_EVENT"]==0])
female_not = len(female[df["DEATH_EVENT"]==1])

labels = 'Male Survived', 'Male Not Survived', 'Female Survived', 'Female Not Survived'
size = [male_survived, male_not, female_survived, female_not]
explode = (0, 0.1, 0, 0.1)

plt.pie(size, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
# Gender Factor Analysis
survived_age = df[df["DEATH_EVENT"]==0]["age"]
notsurv_age = df[df["DEATH_EVENT"]==1]["age"]
sns.distplot( survived_age , color="skyblue", label="survived", bins=50)
sns.distplot( notsurv_age , color="red", label="not survived", bins=50)
plt.legend()
```

- Survival is high on 40 to 70
<br>
- The spread of Not survival is going through all ages

- Survival spread is high in age's flow of 40 to 70
<br>
- The Survival is high for both male between 50 to 60 and female's age between 60 to 70 respectively

```python
# Age vs Sex Survival Status
Box_AgeSex_survstatus = sns.boxplot(x="sex", y="age", hue="DEATH_EVENT", data=df)
```

- Survival spread is high in age's flow of 40 to 70
<br>
- The Survival is high for both male between 50 to 60 and female's age between 60 to 70 respectively

```python
# Age vs Smoking Survival Status
Box_AgeSmoke_survstatus = sns.boxplot(x="smoking", y="age", hue="DEATH_EVENT", data=df)
```

- The Survival is high for not smoking person 55 to 65, while for smoking person it is between 50 to 60
<br>
- Death event for smoking person is high than not smoking person

```python
# Age vs Diabetes Survival Status
Box_AgeDbt_survstatus = sns.boxplot(x="diabetes", y="age", hue="DEATH_EVENT", data=df)
```
